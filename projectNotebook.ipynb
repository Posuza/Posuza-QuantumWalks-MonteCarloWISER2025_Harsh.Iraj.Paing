````
# Unified QGB generator
from qiskit import QuantumCircuit

def build_galton(n_layers, method='copy', coin_angles=None, measure=True):
    """
    Build a Galton board circuit.
    - n_layers: number of layers (int)
    - method: 'copy' (coin + CNOT->position) or 'peg' (controlled-SWAP cascade)
    - coin_angles: None or scalar/list. If None -> Hadamard (unbiased).
    - measure: whether to add measurement on the position qubits.
    Returns: QuantumCircuit
    """
    if coin_angles is None:
        theta_list = [None]*n_layers
    elif np.isscalar(coin_angles):
        theta_list = [float(coin_angles)]*n_layers
    else:
        theta_list = list(coin_angles)
        assert len(theta_list)==n_layers

    if method == 'copy':
        # coin reused + position qubits
        qc = QuantumCircuit(n_layers+1, n_layers)  # coin + pos | measure pos into n classical bits
        for layer in range(n_layers):
            th = theta_list[layer]
            if th is None:
                qc.h(0)
            else:
                qc.ry(th, 0)
            qc.cx(0, layer+1)
        if measure:
            qc.measure(range(1, n_layers+1), range(n_layers))
        return qc

    elif method == 'peg':
        # coin at qubit 0, position qubits 1..n_layers
        qc = QuantumCircuit(n_layers+1, n_layers)
        qc.x(1)  # init walker at leftmost bucket
        for layer in range(n_layers):
            th = theta_list[layer]
            if th is None:
                qc.h(0)
            else:
                qc.ry(th, 0)
            for k in range(layer+1):
                qc.cswap(0, 1+k, 1+k+1)
            # optional cleanup
        if measure:
            qc.measure(range(1, n_layers+1), range(n_layers))
        return qc
    else:
        raise ValueError("method must be 'copy' or 'peg'")
````

````
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

def gaussian_approx(n_layers):
    mu = n_layers/2
    var = n_layers/4
    x = np.arange(n_layers+1)
    g = np.exp(-0.5 * ((x-mu)**2)/var) / np.sqrt(2*pi*var)
    return g / g.sum()

def statevector_bin_probs(qc):
    sv = Statevector.from_instruction(qc.remove_final_measurements(inplace=False))
    # map basis -> bins (# of ones in position qubits)
    n_qubits = qc.num_qubits
    pos_start = 1  # our convention
    n_positions = qc.num_qubits - 1
    probs = np.abs(sv.data)**2
    bins = np.zeros(n_positions+1)
    for idx,p in enumerate(probs):
        if p==0: continue
        bits = format(idx, '0{}b'.format(n_qubits))
        ones = 0
        for pos in range(n_positions):
            q_idx = pos_start + pos
            char_index = n_qubits - 1 - q_idx
            if bits[char_index]=='1':
                ones += 1
        bins[ones] += p
    return bins / bins.sum()

sim = AerSimulator()
for n in [1,2,3,4,6,8]:
    qc = build_galton(n, method='copy', measure=False)  # statevector mode; no measurement
    probs = statevector_bin_probs(qc)
    gauss = gaussian_approx(n)
    plt.figure(figsize=(5,3))
    plt.bar(range(n+1), probs, alpha=0.7, label='statevector')
    plt.plot(range(n+1), gauss, 'k--', label='gaussian approx')
    plt.title(f'n_layers={n}')
    plt.xlabel('bin')
    plt.legend()
    plt.show()
````

**Task 3: Building the Quantum Walk Operators**
This task focuses on implementing the core of the quantum walk-the coin and shift operators.
Code Cells: Cell 7, Cell 13, Cell 17
Referenced Paper Sections:
arxiv:2410.19489: Section 3.3, "Transport," with a focus on the Coin operator and Shift operator6.
arxiv:2202.01735: Section 3.2, "Quantum circuit model of a Markov chain," which provides the theoretical foundation for modeling stochastic processes like particle transport77.


**Implementation Details:**
The code in these cells defines the coin operator, which uses
 position qubits as controls to apply rotations to the coin qubits8. This is a direct implementation of the "position-dependent coin operator" required for the non-homogeneous bypass geometry9.


The code also includes the
 shift operator logic, which would typically be implemented using the Quantum Fourier Transform (QFT) as a "quantum adder" to update the particle's position based on the coin state10. This modular design directly translates the core algorithmic details of the
arxiv:2410.19489 paper into a functional circuit11.


````
# Unified Quantum Galton Board (QGB) generator
import numpy as np
from qiskit import QuantumCircuit

def _p_to_theta(p):
    """Convert probability p (P(1)=p) to rotation angle theta for Ry/Rx: sin^2(theta/2) = p."""
    p = float(p)
    if p < 0 or p > 1:
        raise ValueError("p must be between 0 and 1.")
    return 2 * np.arcsin(np.sqrt(p))

def generate_galton_circuit(n_layers, p=None, theta=None, coin_mode='single', measure=True, bias_gate='ry'):
    """
    Generate a Galton-board-like quantum circuit (QGB).

    Parameters
    ----------
    n_layers : int
        Number of layers / steps in the Galton board.
    p : float, optional
        If provided, sets the bias so that P(1)=p at each coin toss (theta derived).
    theta : float or list, optional
        If scalar, used for all layers. If list of length n_layers, applies per-layer angles.
        If None and p is None, an unbiased (Hadamard) coin is used.
    coin_mode : {'single', 'per_layer'}
        'single' : one coin qubit reused each layer, copying outcome into a fresh position qubit.
        'per_layer' : one coin qubit per layer, measured directly.
    measure : bool
        If True, measures the position/coin qubits into classical bits.
    bias_gate : {'ry','rx'}
        Rotation gate used for biased coin. P(1)=sin^2(theta/2).

    Returns
    -------
    QuantumCircuit
        The constructed Qiskit QuantumCircuit.
    """
    if p is not None:
        theta_from_p = _p_to_theta(p)
        if theta is None:
            theta = theta_from_p

    if theta is None:
        theta_list = [None] * n_layers
    elif np.isscalar(theta):
        theta_list = [float(theta)] * n_layers
    else:
        theta_list = list(theta)
        if len(theta_list) != n_layers:
            raise ValueError("theta list must have length n_layers")

    if coin_mode == 'single':
        n_qubits = n_layers + 1
        n_clbits = n_layers if measure else 0
        qc = QuantumCircuit(n_qubits, n_clbits)
        for layer in range(n_layers):
            th = theta_list[layer]
            if th is None:
                qc.h(0)
            else:
                getattr(qc, bias_gate.lower())(th, 0)
            qc.cx(0, layer + 1)
        if measure:
            qc.measure(range(1, n_layers + 1), range(n_clbits))
        return qc

    elif coin_mode == 'per_layer':
        n_qubits = n_layers
        n_clbits = n_layers if measure else 0
        qc = QuantumCircuit(n_qubits, n_clbits)
        for q in range(n_layers):
            th = theta_list[q]
            if th is None:
                qc.h(q)
            else:
                getattr(qc, bias_gate.lower())(th, q)
        if measure:
            qc.measure(range(n_qubits), range(n_clbits))
        return qc
    else:
        raise ValueError("coin_mode must be 'single' or 'per_layer'")

# Alias for convenience
QGB = generate_galton_circuit
````

````
# Unified Quantum Galton Board (QGB) generator
import numpy as np
from qiskit import QuantumCircuit

def _p_to_theta(p):
    """Convert probability p (P(1)=p) to rotation angle theta for Ry/Rx: sin^2(theta/2) = p."""
    p = float(p)
    if p < 0 or p > 1:
        raise ValueError("p must be between 0 and 1.")
    return 2 * np.arcsin(np.sqrt(p))

def generate_galton_circuit(n_layers, p=None, theta=None, coin_mode='single', measure=True, bias_gate='ry'):
    """
    Returns a Qiskit circuit that implements a Galton-board-like walk by using
    coin rotations and CX copying into position qubits (simple, efficient).
    coin_mode: 'single' (1 coin reused + position qubits) or 'per_layer' (one coin per layer).
    """
    if p is not None:
        theta_from_p = _p_to_theta(p)
        if theta is None:
            theta = theta_from_p

    if theta is None:
        theta_list = [None] * n_layers
    elif np.isscalar(theta):
        theta_list = [float(theta)] * n_layers
    else:
        theta_list = list(theta)
        if len(theta_list) != n_layers:
            raise ValueError("theta list must have length n_layers")

    if coin_mode == 'single':
        n_qubits = n_layers + 1
        n_clbits = n_layers if measure else 0
        qc = QuantumCircuit(n_qubits, n_clbits)
        for layer in range(n_layers):
            th = theta_list[layer]
            if th is None:
                qc.h(0)
            else:
                getattr(qc, bias_gate.lower())(th, 0)
            qc.cx(0, layer + 1)   # copy coin into position qubit
        if measure:
            qc.measure(range(1, n_layers+1), range(n_clbits))
        return qc

    elif coin_mode == 'per_layer':
        n_qubits = n_layers
        n_clbits = n_layers if measure else 0
        qc = QuantumCircuit(n_qubits, n_clbits)
        for q in range(n_layers):
            th = theta_list[q]
            if th is None:
                qc.h(q)
            else:
                getattr(qc, bias_gate.lower())(th, q)
        if measure:
            qc.measure(range(n_qubits), range(n_clbits))
        return qc
    else:
        raise ValueError("coin_mode must be 'single' or 'per_layer'")

# alias
QGB = generate_galton_circuit
````

````
# Peg-based Galton Board (Fredkin / cswap cascade)
# Idea: positions are qubits q_pos[0..n_layers].  Start with amplitude in q_pos[0].
# For each layer `l` (0..n_layers-1) prepare a coin (H) then for k in 0..l apply controlled-SWAP between q_pos[k] and q_pos[k+1]
# controlled by the coin. This routes amplitude between adjacent buckets, similar to pegs.
def generate_peg_galton_circuit(n_layers, measure=True):
    """
    Peg-based QGB:
      - position qubits: n_layers+1
      - coin qubit reused per layer (qubit index 0), positions start at index 1..n_layers+1
    Returns: QuantumCircuit
    """
    # coin qubit + positions
    n_positions = n_layers + 1
    coin_index = 0
    pos_start = 1
    n_qubits = 1 + n_positions    # 1 coin + positions
    n_clbits = n_positions if measure else 0
    qc = QuantumCircuit(n_qubits, n_clbits)

    # Initialize amplitude: put a |1> in the topmost position (pos 0) to flow downward
    # We'll use pos qubit pos_start (which corresponds to bucket index 0)
    qc.x(pos_start)  # initial excitation in leftmost/top bucket

    # For each layer apply coin (H) and then controlled-swaps between neighboring positions
    for layer in range(n_layers):
        # unbiased coin
        qc.h(coin_index)
        # for each peg in this layer apply cswap between position k and k+1 controlled by coin
        for k in range(layer + 1):
            a = pos_start + k
            b = pos_start + k + 1
            # Qiskit provides cswap(control, a, b)
            qc.cswap(coin_index, a, b)
        # optionally, we could uncompute/reset coin, but we reuse coin by leaving it and applying H again next layer
        # to get independent coin for next layer, apply H to coin (transform basis)
        # However for simplicity we will reprepare coin next iteration by applying H again (which toggles)
        # To ensure coin doesn't entangle previous positions undesirably, apply H then reset via H at end:
        qc.h(coin_index)  # this attempts to return coin to |0> (subject to entanglement)
        # If desired, more advanced reset/uncompute can be used (paper uses ancilla reuse tricks).

    if measure:
        # measure positions only into classical bits (0..n_positions-1)
        qc.measure(range(pos_start, pos_start + n_positions), range(n_clbits))

    return qc
````

````
# Cell A: amplitude flow heatmap & animation for peg circuit
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML, display
from matplotlib import animation
from qiskit.quantum_info import Statevector

# Helper: compute per-position probabilities from a statevector
def position_probabilities_from_statevector(statevector, n_qubits, pos_start, n_positions):
    """
    statevector: 1D complex numpy array or qiskit Statevector
    n_qubits: total number of qubits in the circuit (coin + positions)
    pos_start: index of first position qubit (int)
    n_positions: number of position qubits
    Returns: list of probabilities [P(pos0), P(pos1), ...]
    """
    if hasattr(statevector, "data"):
        vec = np.asarray(statevector.data)
    else:
        vec = np.asarray(statevector)
    probs = np.abs(vec)**2
    # prepare output
    pos_probs = np.zeros(n_positions)
    # iterate basis states
    # For each basis index, decode bits with length n_qubits with LSB as qubit 0 mapping: qiskit uses little-endian in Statevector.from_instruction,
    # but to be robust we check both. We'll assume Qiskit's computational basis ordering where index bits correspond to [q_{n-1}...q_0]
    for idx, p in enumerate(probs):
        if p == 0:
            continue
        bitstr = format(idx, '0{}b'.format(n_qubits))  # MSB = q_{n_qubits-1}
        # qiskit ordering: leftmost char -> highest index qubit.
        # We want access to position qubits located at indices pos_start..pos_start+n_positions-1
        # mapping: char index = (n_qubits - 1 - qubit_index)
        for pos in range(n_positions):
            q_idx = pos_start + pos
            char_index = n_qubits - 1 - q_idx
            if bitstr[char_index] == '1':
                pos_probs[pos] += p
    return pos_probs

# Build per-layer statevectors for peg circuit up to each layer
def amplitude_flow_for_peg(n_layers):
    """
    Returns array of shape (n_layers+1, n_positions) with probabilities per layer.
    Layer 0 = after initialization only (before any coin/peg),
    Layer k = after applying k layers.
    """
    n_positions = n_layers + 1
    pos_start = 1
    coin_index = 0
    n_qubits = 1 + n_positions
    flow = []
    # Build circuits up to each layer and get statevector
    for layer in range(0, n_layers + 1):
        # create a short peg-circuit up to 'layer' layers
        from qiskit import QuantumCircuit
        qc = QuantumCircuit(n_qubits)
        # initialize particle in leftmost/top bucket
        qc.x(pos_start)
        for L in range(layer):
            # coin H
            qc.h(coin_index)
            # cswaps for k in [0..L]
            for k in range(L + 1):
                a = pos_start + k
                b = pos_start + k + 1
                qc.cswap(coin_index, a, b)
            # apply H again to attempt to reset coin (paper has different approaches)
            qc.h(coin_index)
        # get statevector
        sv = Statevector.from_instruction(qc)
        pos_probs = position_probabilities_from_statevector(sv, n_qubits, pos_start, n_positions)
        flow.append(pos_probs)
    return np.array(flow)  # shape (n_layers+1, n_positions)

# Plot static heatmap and create animation
def plot_and_animate_flow(n_layers, cmap='viridis', save_mp4=True, mp4_name='peg_flow.mp4'):
    flow = amplitude_flow_for_peg(n_layers)  # (layers+1, positions)
    layers = np.arange(flow.shape[0])
    positions = np.arange(flow.shape[1])

    # Static heatmap
    plt.figure(figsize=(8,4))
    plt.imshow(flow, aspect='auto', origin='lower', cmap=cmap)
    plt.colorbar(label='Probability')
    plt.xlabel('Position index')
    plt.ylabel('Layer (0 = init)')
    plt.title(f'Amplitude flow heatmap (peg-based) n_layers={n_layers}')
    plt.xticks(positions)
    plt.yticks(layers)
    plt.show()

    # Animation: each frame is a bar plot of position probs at that layer
    fig, ax = plt.subplots(figsize=(7,3.5))
    bar_container = ax.bar(positions, flow[0], align='center', alpha=0.8)
    ax.set_ylim(0, max(0.6, flow.max()*1.1))
    ax.set_xlabel('Position')
    ax.set_ylabel('Probability')
    ax.set_title('Amplitude flow (per layer)')

    def update(frame):
        for rect, h in zip(bar_container, flow[frame]):
            rect.set_height(h)
        ax.set_title(f'Layer {frame} / {n_layers}')
        return bar_container

    anim = animation.FuncAnimation(fig, update, frames=flow.shape[0], interval=900, blit=False)

    # Save MP4 (requires ffmpeg available in Colab)
    if save_mp4:
        try:
            anim.save(mp4_name, writer='ffmpeg', fps=1.2)
            print(f"Saved animation to {mp4_name}")
        except Exception as e:
            print("Could not save MP4 (ffmpeg missing?). Error:", e)
    plt.close(fig)

    # display inline
    try:
        from IPython.display import HTML
        html = anim.to_jshtml()
        display(HTML(html))
    except Exception:
        # fallback: display saved mp4 if present
        import os
        if os.path.exists(mp4_name):
            display(HTML(f'<video controls src="{mp4_name}" />'))
        else:
            print("Animation could not be displayed inline; try enabling ffmpeg or use the saved mp4.")

# Example usage (change n_layers as desired)
n_layers = 5
plot_and_animate_flow(n_layers, save_mp4=True, mp4_name=f'peg_flow_n{n_layers}.mp4')
````

````
# Cell C: VQA-style variational coin + optimizer
import numpy as np
import matplotlib.pyplot as plt
from qiskit.quantum_info import Statevector
from math import comb
from functools import partial

# Builds peg circuit using supplied coin angles (theta list), no measurement, returns Statevector
def peg_circuit_statevector_with_angles(n_layers, angles):
    n_positions = n_layers + 1
    pos_start = 1
    coin_index = 0
    n_qubits = 1 + n_positions
    from qiskit import QuantumCircuit
    qc = QuantumCircuit(n_qubits)
    qc.x(pos_start)  # initialize particle in leftmost bucket
    for L in range(n_layers):
        theta = angles[L]
        # variational rotation (Ry) on coin
        qc.ry(theta, coin_index)
        # controlled-swaps
        for k in range(L + 1):
            a = pos_start + k
            b = pos_start + k + 1
            qc.cswap(coin_index, a, b)
        # optional uncompute: apply Ry(-theta) or H depending on design; here we keep simple
    # return statevector
    sv = Statevector.from_instruction(qc)
    return sv

# Compute distribution on positions given statevector
def peg_statevector_to_position_distribution(sv, n_layers):
    n_positions = n_layers + 1
    pos_start = 1
    n_qubits = 1 + n_positions
    probs = position_probabilities_from_statevector(sv, n_qubits, pos_start, n_positions)
    # convert to dict keyed 0..n_positions-1
    return {i: float(probs[i]) for i in range(n_positions)}

# Target (binomial)
def binomial_target(n_layers):
    n = n_layers
    return {k: comb(n, k) * (0.5**n) for k in range(n+1)}

# objective: TV distance between distribution from angles and target (binomial)
def objective_tv_distance(angles, n_layers, target):
    sv = peg_circuit_statevector_with_angles(n_layers, angles)
    dist = peg_statevector_to_position_distribution(sv, n_layers)
    # compute TVD
    tv = 0.5 * sum(abs(dist.get(k,0) - target.get(k,0)) for k in set(dist.keys()).union(target.keys()))
    return float(tv)

# Optimizer wrapper: try scipy minimize Nelder-Mead, else random search
def optimize_angles(n_layers, init_angles=None, maxiter=150, random_trials=300):
    if init_angles is None:
        init_angles = np.random.uniform(0, 2*np.pi, size=n_layers)
    target = binomial_target(n_layers)
    try:
        from scipy.optimize import minimize
        res = minimize(lambda x: objective_tv_distance(x, n_layers, target),
                       x0=init_angles, method='Nelder-Mead',
                       options={'maxiter': maxiter, 'disp': True})
        best_angles = res.x
        best_val = res.fun
        return best_angles, best_val, 'scipy_nm'
    except Exception as e:
        print("SciPy minimize unavailable or failed:", e)
        # simple random search fallback
        best_angles = init_angles.copy()
        best_val = objective_tv_distance(best_angles, n_layers, target)
        for t in range(random_trials):
            cand = np.random.uniform(0, 2*np.pi, size=n_layers)
            val = objective_tv_distance(cand, n_layers, target)
            if val < best_val:
                best_val = val
                best_angles = cand
        return best_angles, best_val, 'random_search'

# Wrapper: run optimization and plot before/after + save slides
def run_vqa_and_export(n_layers=4, shots=None, save_prefix='vqa'):
    print(f"Running VQA-style optimization for n_layers={n_layers}")
    target = binomial_target(n_layers)
    # initial angles
    init_angles = np.random.uniform(0, 2*np.pi, size=n_layers)
    sv_init = peg_circuit_statevector_with_angles(n_layers, init_angles)
    dist_init = peg_statevector_to_position_distribution(sv_init, n_layers)
    # optimize
    best_angles, best_val, method = optimize_angles(n_layers, init_angles=init_angles)
    sv_best = peg_circuit_statevector_with_angles(n_layers, best_angles)
    dist_best = peg_statevector_to_position_distribution(sv_best, n_layers)

    # Plot before/after vs target
    positions = list(sorted(dist_init.keys()))
    x = np.arange(len(positions))

    plt.figure(figsize=(9,4))
    plt.subplot(1,2,1)
    plt.bar(x, [dist_init[k] for k in positions], alpha=0.8, label='Init')
    plt.plot(positions, [target[k] for k in positions], 'k--', label='Target (binomial)')
    plt.title('Before optimization')
    plt.xticks(x, positions)
    plt.legend()

    plt.subplot(1,2,2)
    plt.bar(x, [dist_best[k] for k in positions], alpha=0.8, label='Optimized')
    plt.plot(positions, [target[k] for k in positions], 'k--', label='Target (binomial)')
    plt.title(f'After optimization (method={method}, TVD={best_val:.4f})')
    plt.xticks(x, positions)
    plt.legend()

    plt.suptitle(f'VQA tuning of peg coin (n_layers={n_layers})')
    plt.tight_layout(rect=[0,0.03,1,0.95])
    # Save slide figure
    caption = f"VQA run: n_layers={n_layers}, optimizer={method}, final TV distance={best_val:.4f}"
    save_slide_figure(caption_text=caption, filename_prefix=save_prefix)
    plt.show()

    print("Best TV distance:", best_val)
    print("Best angles (radians):", best_angles)
    return best_angles, best_val, dist_init, dist_best

# Example run (run this cell; optimization may take time depending on n_layers)
n_layers_vqa = 4
best_angles, best_val, dist_init, dist_best = run_vqa_and_export(n_layers_vqa, save_prefix=f'vqa_n{n_layers_vqa}')
````

````
# ===== Cell 1: imports & helper utilities (run once) =====
# Assumes earlier code: generate_peg_galton_circuit, generate_galton_circuit,
# counts_to_position_distribution, position_probabilities_from_statevector, save_slide_figure exist.

import numpy as np
import matplotlib.pyplot as plt
from math import comb
from IPython.display import display, Markdown, clear_output
import ipywidgets as widgets
from qiskit import transpile
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
import time

sim = AerSimulator()

# Theoretical targets
def binomial_target(n_layers):
    return {k: comb(n_layers, k) * (0.5**n_layers) for k in range(n_layers+1)}

def tv_distance(distA, distB):
    keys = sorted(set(distA.keys()).union(distB.keys()))
    return 0.5 * sum(abs(distA.get(k,0) - distB.get(k,0)) for k in keys)

# Short plotting helper
def plot_sim_vs_target(dist_sim, target, title="Sim vs Target"):
    ks = sorted(dist_sim.keys())
    vals_sim = [dist_sim[k] for k in ks]
    vals_t = [target.get(k,0) for k in ks]
    x = np.arange(len(ks))
    plt.figure(figsize=(7,3.4))
    plt.bar(x-0.2, vals_sim, width=0.4, label='Sim', alpha=0.85)
    plt.plot(x+0.2, vals_t, 'k--', marker='o', label='Target')
    plt.xticks(x, ks)
    plt.xlabel('Bin (number of 1s)')
    plt.ylabel('Probability')
    plt.title(title)
    plt.legend()
    plt.show()

# A robust simulate-measured function (run shots, return distribution dict keyed 0..n_positions-1)
def run_shot_simulation(qc, shots=1024):
    """Transpile and run qc on AerSimulator with given shots. Returns position distribution dict."""
    compiled = transpile(qc, sim)
    job = sim.run(compiled, shots=shots)
    res = job.result()
    counts = res.get_counts()
    # convert counts: number of '1's in bitstring (measured bits) -> bin
    n_pos = qc.num_clbits if qc.num_clbits>0 else qc.num_qubits-1
    dist = counts_to_position_distribution(counts, n_positions=n_pos)
    return dist, counts

# Quick statevector-run for circuits without measurements
def run_statevector_qc(qc):
    sv = Statevector.from_instruction(qc)
    return sv

# Decision metric for "interestingness": TV distance to binomial target (higher => more quantum/unexpected)
def interestingness_metric(dist_sim, n_layers):
    t = binomial_target(n_layers)
    return tv_distance(dist_sim, t)

# Dynamic allocation policy: initial low_shots, evaluate metric, if metric>threshold request extra shots up to max_shots
def dynamic_allocation_runner(qc_builder, n_layers, initial_shots=200, threshold=0.12, max_shots=5000):
    """
    qc_builder: function that returns a measured QuantumCircuit for given n_layers
    returns: final_dist, history list of (shots, dist, metric)
    """
    shots = initial_shots
    history = []
    qc = qc_builder(n_layers)
    dist, raw_counts = run_shot_simulation(qc, shots=shots)
    metric = interestingness_metric(dist, n_layers)
    history.append((shots, dist, metric))
    # Keep doubling shots while metric > threshold and shots < max_shots
    while metric > threshold and shots < max_shots:
        shots = min(shots*2, max_shots)
        dist, raw_counts = run_shot_simulation(qc, shots=shots)
        metric = interestingness_metric(dist, n_layers)
        history.append((shots, dist, metric))
        if len(history) > 6:  # safety
            break
    return dist, history
````

````
# Cell B — Peg Galton Board Builder
def build_galton_circuit(n_layers, coin_angle=np.pi/4, measure=True):
    """Creates a quantum Galton board circuit with n_layers and adjustable coin rotation."""
    n_qubits = n_layers + 1
    qc = QuantumCircuit(n_qubits)

    # Start all walkers in |0>
    for layer in range(n_layers):
        qc.ry(coin_angle, layer)
        qc.cx(layer, layer+1)

    if measure:
        qc.measure_all()
    return qc
````

````
# Cell D — Compute L1 distance between distributions
def l1_distance(counts_a, counts_b):
    keys = set(counts_a.keys()).union(counts_b.keys())
    total_shots_a = sum(counts_a.values())
    total_shots_b = sum(counts_b.values())
    dist = 0
    for k in keys:
        pa = counts_a.get(k, 0) / total_shots_a
        pb = counts_b.get(k, 0) / total_shots_b
        dist += abs(pa - pb)
    return dist

dist = l1_distance(counts_noiseless, counts_noisy)
print(f"L1 distance between noisy and noiseless: {dist:.4f}")
````
